```python
from huggingface_hub import list_repo_files

dataset_id = "EuroBERT/the-stack-v2"
subpath = "data/Python"

all_files = list_repo_files(dataset_id, repo_type="dataset")
```

```python
shuffled_files = all_files.copy()
import random
random.shuffle(shuffled_files)
```

```python
by_lang = {
}
for item in all_files:
    try:
        lang = item.split('/')[1]
        if lang not in by_lang:
            by_lang[lang] = []
        by_lang[lang].append(item)
    except IndexError:
        print(f"Error processing item: {item}")
```

```python
filter_pretrained = {
    "HTML": 0.03,
    "JSON": 0.01,
    "YAML": 0.1,
    "Markdown": 0.05,
}
```

```python
filter_1 = by_lang.copy()
for lang in by_lang:
    filter_1[lang] = filter_1[lang][:int(len(by_lang[lang]) * filter_pretrained.get(lang, 1))]
    print(f"Filtered {lang}: {len(by_lang[lang])} -> {len(filter_1[lang])}")
```

```python
filter_annealing = {
    "C++": 0.25,
    "C": 0.5,
    "HTML": 0.01,
    "JavaScript": 0.5,
    "JSON": 0.01,
    "PHP": 0.6,
    "SQL": 0.25,
    "YAML": 0.05,
}
```

```python
filter_2 = filter_1.copy()
for lang in filter_1:
    filter_2[lang] = filter_2[lang][:int(len(by_lang[lang]) * filter_annealing.get(lang, 1))]
    print(f"Filtered {lang}: {len(filter_1[lang])} -> {len(filter_2[lang])}")
```

```python
final_filter = filter_2.copy()
for lang in filter_2:
    final_filter[lang] = final_filter[lang][:int(len(by_lang[lang]) * 0.1)]
    print(f"Filtered {lang}: {len(by_lang[lang])} -> {len(final_filter[lang])}")
```

```python
all_files = [item for sublist in final_filter.values() for item in sublist]
```
